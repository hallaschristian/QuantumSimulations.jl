# Force curves versus velocity for an $F_1 \rightarrow F_1$ system
In this notebook, we simulate cooling for an $F_1 \rightarrow F_1$ system. In particular, we reproduce figures from [NJP 18, 123017 (2016)](https://doi.org/10.1088/1367-2630/18/12/123017).

```julia
using Revise
```

```julia
using
    QuantumStates,
    OpticalBlochEquations,
    DifferentialEquations,
    UnitsToValue
;
```

We'll first define a few physical constants needed for this simulation. 

```julia
λ = 589e-9
Γ = 2π * 9.8e6
m = @with_unit 23 "u"
k = 2π / λ
;
```

Using `QuantumStates`, we can create both the ground states (`F1_lower`) and excited states (`F1_upper`) using the `AngularMomentumState` type, which is a state type of "bare" angular momentum states $|F,m\rangle$.

```julia
# F = 1
QN_bounds = (E = 0.0, F = 1)
F1_lower = enumerate_states(AngularMomentumState, QN_bounds)
QN_bounds = (E = c / λ, F = 1)
F1_upper = enumerate_states(AngularMomentumState, QN_bounds)

ground_states = F1_lower
excited_states = F1_upper
states = [ground_states; excited_states]
;
```

The transition dipole moments `d` and the magnetic moments `d_m` are calculated using `QuantumStates.get_tdms_two_bases`:

```julia
d_ge = get_tdms_two_bases(ground_states, excited_states, TDM)
d = [
    [zeros(length(ground_states), length(excited_states), 3); permutedims(d_ge, (2,1,3))] [d_ge; zeros(length(excited_states), length(ground_states), 3)]
]

d_m_ge = get_tdms_two_bases(ground_states, excited_states, TDM_magnetic)
d_m = [
    [d_m_ge; zeros(length(ground_states), length(excited_states), 3)] [zeros(length(excited_states), length(ground_states), 3); d_m_ge]
]
;
```

We now define the lasers, along with the detuning $\Delta$ and saturation $s$. Note that all lasers have the polarization $\sigma^+$ in their own frame, which is rotated to the $\hat{z}$ axis using `rotate_pol`. (Technically, the rotation is performed _from_ the $\hat{z}$ axis _to_ the axis of the given laser's $k$-vector because the variable $\sigma^+$ is defined relative to the $\hat{z}$ axis.)

```julia
# Laser parameters
Δ = -Γ
s = 1.0

# Frequency of the lasers (in angular frequency units)
ω_F1_to_F1 = 2π * (F1_upper[1].E - F1_lower[1].E) + Δ

τ = 0.25e-6 #/ 2π
const f_RF = round(2π / (Γ * τ), digits=1)
const _μB = (μ_B / h) * (1e-6 * 1e-4)
ϵ(ϵ1, ϵ2) = t -> Bool(mod(div(f_RF * t, 2π), 2)) ? ϵ1 : ϵ2
# ϵ(ϵ1, ϵ2) = t -> ϵ1

k̂ = +x̂; ϵ1 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser1 = Field(k̂, ϵ1, ω_F1_to_F1, s)
k̂ = -x̂; ϵ2 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser2 = Field(k̂, ϵ2, ω_F1_to_F1, s)
k̂ = +ŷ; ϵ3 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser3 = Field(k̂, ϵ3, ω_F1_to_F1, s)
k̂ = -ŷ; ϵ4 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser4 = Field(k̂, ϵ4, ω_F1_to_F1, s)
k̂ = +ẑ; ϵ5 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser5 = Field(k̂, ϵ5, ω_F1_to_F1, s)
k̂ = -ẑ; ϵ6 = ϵ(rotate_pol(σ⁺, k̂), rotate_pol(σ⁻, k̂)); laser6 = Field(k̂, ϵ6, ω_F1_to_F1, s)

lasers = [laser1, laser2, laser3, laser4, laser5, laser6]
;
```

```julia
# The index mapping of states at positive B to negative B 
states_correspondence = [
    3
    2
    1
    6
    5
    4
];
```

```julia
function flip_Bfield!(integrator)
    # "Flip" the energy of the states diabatically
    integrator.p.B .*= -1
    # integrator.p.ω = integrator.p.ω[integrator.p.extra_p.states_correspondence]
    return nothing
end
flip_Bfield_cb = PeriodicCallback(flip_Bfield!, 2π / f_RF)
;
```

Before computing the force across a range of velocities, let's first check that our simulation produces reasonable results for a specific set of parameters:

```julia
# `Particle` type defines the starting position `r0` and velocity `v` used for the simulation
particle = Particle()
particle.r0 = [0.0, 0.0, 0.0]
particle.v = [0.0, 0.0, 0.0]

# Define the density matrix `ρ0` at t = 0
ρ0 = zeros(ComplexF64, length(states), length(states))
ρ0[1,1] = 1.0

# `freq_res` designates the resolution used for the frequencies and velocity used in the simulation
freq_res = 1e-1

extra_p = (states_correspondence=states_correspondence,)
p = obe(ρ0, particle, states, lasers, d, d_m, true, true, λ, Γ, freq_res, extra_p)
prob = ODEProblem(ρ!, p.ρ0_vec, tspan, p) 

# B = (0.0, 0.0, 0.0)
prob.p.v = (0.0, 0.0, 0.0)
# prob.p.B = B
# prob.p.ω += ((2π * B[3]) .* diag(d_m[:,:,2]))

# Define the end time `t_end` of the simulation
t_end = 30p.period+1
tspan = (0., t_end)
;
```

```julia
force_callback = PeriodicCallback(reset_force!, prob.p.period)
cbs = CallbackSet(force_callback, flip_Bfield_cb)

@time sol = DifferentialEquations.solve(prob, alg=DP5(), callback=cbs, abstol=1e-5);
```

```julia
sol.t[end], t_end
```

```julia
using Plots
plot_us = sol.u
plot_ts = sol.t

n_states = size(p.ρ_soa, 1)
plot(size=(800, 400), ylim=(-0.1, 1.1), legend=nothing)
for i in 1:n_states
    state_idx = n_states*(i-1) + i
    plot!(plot_ts, [real(u[state_idx]) for u in plot_us], linewidth=2)
end
plot!()
offset = 20
vline!([sol.t[end] - p.period - offset, sol.t[end] - offset], color="red", linestyle=:dash)
```

## Laser cooling force versus velocity

```julia
using Distributions
uniform_dist = Uniform(0, 2π)
function sample_direction(r=1.0)
    θ = 2π * rand()
    z = rand() * 2 - 1
    return (r * sqrt(1 - z^2) * cos(θ), r * sqrt(1 - z^2) * sin(θ), r * z)
end
;
```

```julia
function prob_func!(prob, scan_values, i)
    # Update velocity and position
    prob.p.v .= (scan_values.v[i], 0.0, 0.1) #sample_direction(inner_config.v[i])
    prob.p.r0 .= rand(uniform_dist, 3)
    return prob
end
function param_func(p, scan_values, i)
    return round(norm(p.v), digits=3)
end
function output_func(p, sol)
    f = p.force_last_period[1]
    return f
end
;
```

```julia
t_end = 40p.period+1
tspan = (0., t_end)
prob = ODEProblem(ρ!, p.ρ0_vec, tspan, p, save_on=false, callback=flip_Bfield_cb) 

vs = repeat(0:0.1:8, 100)
scan_values = (v = vs,)
@time scan_params, forces = force_scan(prob, scan_values, prob_func!, param_func, output_func);
```

```julia
vs, averaged_forces = average_forces(scan_params, forces)
plot(vs, 1e3 .* averaged_forces, legend=nothing, linewidth=2, size=(450,400))
```

## Laser cooling force versus magnetic field

```julia
function prob_func!(prob, scan_values, i)
    p = prob.p
    # Update velocity and position
    prob.p.v .= sample_direction(scan_values.vs[i])
    prob.p.r0 .= rand(uniform_dist, 3)
    # Round `v` to ensure that the OBEs are periodic
    prob.p.v .= round_vel(prob.p.v, prob.p.freq_res)
    prob.p.B .= (0.0, 0.0, scan_values.Bz[i])
    # prob.p.ω .= p.extra_p.ω0 + (2π * scan_values.Bz[i]) .* diag(p.d_m[:,:,2])
    return prob
end
function param_func(p, scan_values, i)
    return scan_values.Bz[i]
end
function output_func(p, sol)
    f = p.force_last_period[3]
    return f
    # f = p.force_last_period
    # return f ⋅ p.v / norm(p.v)
end
;
```

```julia
extra_p = (states_correspondence=states_correspondence, ω0=prob.p.ω)
p = obe(ρ0, particle, states, lasers, d, d_m, true, true, λ, Γ, freq_res, extra_p)
prob = ODEProblem(ρ!, p.ρ0_vec, tspan, p, save_on=false, callback=flip_Bfield_cb)

n_samples = 500

Bzs = (0:3:30.0) ./ (Γ / (_μB * 1e6))
vs  = 0.1 .* ones(length(Bzs))
scan_values = (Bz = repeat(Bzs, n_samples), vs = repeat(vs, n_samples))

@time scan_params, forces = force_scan(prob, scan_values, prob_func!, param_func, output_func);
```

```julia
using LaTeXStrings
Bzs, averaged_forces = average_forces(scan_params, forces)
plot(Bzs .* (Γ / (_μB * 1e6)), (1e-3 * ħ * k * Γ / m) .* averaged_forces, 
    xlabel="Magnetic field along z (ħΓ/μ)",
    ylabel="z force (10⁻³ ħkΓ)",
    legend=nothing, 
    linewidth=2)
```
