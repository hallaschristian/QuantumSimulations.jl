```python
import pyvisa
import time
import matplotlib.pyplot as plt
from ThorlabsPM100 import ThorlabsPM100
rm = pyvisa.ResourceManager()
inst = rm.open_resource('USB0::0x1313::0x8078::P0008588::INSTR',
                         timeout=1)
power_meter = ThorlabsPM100(inst=inst)
```

```python
out=[]
for i in range(100):
    time.sleep(0.05)
    out.append(power_meter.read)
plt.plot(out)
```

```python
inst.close()
```
