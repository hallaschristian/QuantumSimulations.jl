```julia
@everywhere using
    QuantumStates,
    OpticalBlochEquations,
    DifferentialEquations,
    UnitsToValue,
    LinearAlgebra,
    Printf,
    Plots,
    Random,
    StatsBase,
    Distributions,
    StructArrays,
    StaticArrays,
    StructArrays,
    LoopVectorization,
    Parameters,
    MutableNamedTuples

@everywhere @consts begin
    λ = 780.241209686e-9
    Γ = 2π * 6.0666e6
    m = @with_unit 87 "u"
    k = 2π / λ
    _μB = (μ_B / h) * 1e-4
    kB = 1.380649e-23
end
;
```

```julia
@everywhere sim_params = MutableNamedTuple(
    grad = 10., # G/cm
    tweezer_intensity = 2*0.001 / (π * (3e-6)^2), # 1 mW tweezer with 3 um waist
    tweezer_size = (3e-6, 35e-6, 3e-6), 
    tweezer_pol = [0,1,0], # linearly polarized tweezer
    tweezer_x_positions = (-10:10) .* 5e-6, # grid to define tweezer array x positions
    tweezer_z_positions = (-10:10) .* 5e-6 # grid to define tweezer array z positions
)
;
```

### Calculate transition dipole moments

```julia
@everywhere begin
    include("define_Rb87_states.jl")
    ground_states, excited_states = define_Rb87_states()
    states = [ground_states; excited_states]
    d = tdms_between_states(states, states)

    # ground states to use for the simulation, F=2 levels in 5^2S_{1/2}
    g_idxs = 4:8

    # excited states to use for the simulation, F=3 levels in 5^P_{3/2}
    e_idxs = 18:24

    idxs = [g_idxs; e_idxs]
    ground_states_to_use = states[g_idxs]
    excited_states_to_use = states[e_idxs]
    states_to_use = states[idxs]
    d_to_use = d[idxs,idxs,:]
end
;
```

### Run simulation

```julia
@everywhere function update_H_and_∇H(H, p, r, t)

    x,y,z = r
    
    bfield_x = x * (-p.sim_params.grad / k) * 1e2 / 2
    bfield_y = y * (+p.sim_params.grad / k) * 1e2 / 2
    bfield_z = z * (-p.sim_params.grad / k) * 1e2
    
    # add Zeeman terms to the Hamiltonian
    @turbo for i ∈ eachindex(H)
        H.re[i] = bfield_x * p.extra_data.zeeman_x_mat.re[i] + bfield_y * p.extra_data.zeeman_y_mat.re[i] + bfield_z * p.extra_data.zeeman_z_mat.re[i]
        H.im[i] = bfield_x * p.extra_data.zeeman_x_mat.im[i] + bfield_y * p.extra_data.zeeman_y_mat.im[i] + bfield_z * p.extra_data.zeeman_z_mat.im[i]
    end
    
    # add tweezer-atom terms to the Hamiltonian
    tweezer_size = p.sim_params.tweezer_size .* p.k
    
    total_tweezer_scalar = 0.0
    ∇H_x = 0.0
    ∇H_y = 0.0
    ∇H_z = 0.0
    
    for tweezer_x ∈ p.sim_params.tweezer_x_positions
        tweezer_x *= p.k
        for tweezer_z ∈ p.sim_params.tweezer_z_positions
            tweezer_z *= p.k
            
            tweezer_scalar = exp(-2(x-tweezer_x)^2/tweezer_size[1]^2) * exp(-2y^2/tweezer_size[2]^2) * exp(-2(z-tweezer_z)^2/tweezer_size[3]^2)
            
            ∇H_x += (-4*(x-tweezer_x) / tweezer_size[1]^2) * tweezer_scalar
            ∇H_y += (-4y / tweezer_size[2]^2) * tweezer_scalar
            ∇H_z += (-4*(z-tweezer_z) / tweezer_size[3]^2) * tweezer_scalar

            total_tweezer_scalar += tweezer_scalar    
        end
    end
         
    @turbo for i in eachindex(H)
        H.re[i] += p.extra_data.H_tweezer.re[i] * total_tweezer_scalar
        H.im[i] += p.extra_data.H_tweezer.im[i] * total_tweezer_scalar
    end
    
    ∇H = SVector{3, Float64}(∇H_x, ∇H_y, ∇H_z)

    return ∇H
end
;
```

```julia
@everywhere extra_data = MutableNamedTuple(
    zeeman_x_mat=zeeman_x_mat,
    zeeman_y_mat=zeeman_y_mat,
    zeeman_z_mat=zeeman_z_mat,
    H_tweezer=H_tweezer
)
;
```

```julia
@everywhere begin
    t_start = 0.0
    t_end   = 15e-3
    t_span  = (t_start, t_end) ./ (1/Γ)

    n_states = length(states_to_use)
    n_excited = length(excited_states_to_use)

    particle = Particle()
    particle.r = (0e-3, 0e-3, 1e-3) ./ (1/k)
    particle.v = (0, 0, 0) ./ (Γ/k)

    ψ₀ = zeros(ComplexF64, n_states)
    ψ₀[1] = 1.0
end

using Logging: global_logger
using TerminalLoggers: TerminalLogger
global_logger(TerminalLogger())

p = schrodinger_stochastic(particle, states_to_use, lasers, d_to_use, ψ₀, m/(ħ*k^2/Γ), n_excited; sim_params=sim_params, extra_data=extra_data, λ=λ, Γ=Γ, update_H_and_∇H=update_H_and_∇H)

prob = ODEProblem(ψ_stochastic_potential!, p.ψ, t_span, p)

cb = ContinuousCallback(condition, SE_collapse_pol_always!, nothing, save_positions=(false,false))

@time sol = DifferentialEquations.solve(prob, alg=DP5(), reltol=5e-4, callback=cb, saveat=100, maxiters=80000000, progress=true, progress_steps=200000)
;
```

```julia
plot_us = sol.u
plot_ts = sol.t

times = sol.t .* (1e3/Γ)

x_trajectories = [real(u[n_states + n_excited + 1]) for u in plot_us] .* (1e3/k)
y_trajectories = [real(u[n_states + n_excited + 2]) for u in plot_us] .* (1e3/k)
z_trajectories = [real(u[n_states + n_excited + 3]) for u in plot_us] .* (1e3/k)

x_velocities = [real(u[n_states + n_excited + 4]) for u in plot_us] ./ (k/Γ)
y_velocities = [real(u[n_states + n_excited + 5]) for u in plot_us] ./ (k/Γ)
z_velocities = [real(u[n_states + n_excited + 6]) for u in plot_us] ./ (k/Γ)
;
```

```julia
plot(times, z_trajectories, xlabel="Time (ms)")
```

```julia
plot(sol.t .* (1e3/Γ), z_velocities,
    xlabel="Time (ms)",
    ylabel="Velocity along z (m/s)"
)
```

### Run simulation for multiple particles in parallel

```julia
@everywhere function prob_func(prob, i, repeat)
    
    lasers = define_lasers(states)
    
    # Define initial conditions for the molecule
    cloud_size = @with_unit 0.025 "mm"
    particle = Particle()
    particle.r = (rand(Normal(0, cloud_size)), rand(Normal(0, cloud_size)), rand(Normal(0, cloud_size))) ./ (1/k) 
    # particle.v = (0,0,10) ./ (Γ/k)
    
    ψ₀ = zeros(ComplexF64, n_states)
    ψ₀[1] = 1.0
    
    p = schrodinger_stochastic(particle, states_to_use, lasers, d_to_use, ψ₀, m/(ħ*k^2/Γ), n_excited; sim_params=sim_params, extra_data=extra_data, λ=λ, Γ=Γ, update_H_and_∇H=update_H_and_∇H)
    
    callback = ContinuousCallback(condition, SE_collapse_pol_always!, nothing, save_positions=(false,false))
    
    prob = ODEProblem(ψ_stochastic_potential!, p.ψ, t_span, p, callback=callback, reltol=5e-4, saveat=1000, maxiters=50000000)
    
    return prob
end
;
```

```julia
ensemble_prob = EnsembleProblem(prob; prob_func=prob_func)
;
```

```julia
n_molecules = 20
@time ensemble_sol = solve(ensemble_prob, DP5(), EnsembleDistributed(); trajectories=n_molecules)
;
```

```julia
x_trajectories = Array{Vector{Float64}}(fill([],n_molecules))
y_trajectories = Array{Vector{Float64}}(fill([],n_molecules)) 
z_trajectories = Array{Vector{Float64}}(fill([],n_molecules))
times = Array{Vector{Float64}}(fill([],n_molecules))

for i ∈ 1:n_molecules
    sol_u = ensemble_sol[i].u
    sol_t = ensemble_sol[i].t
    
    x_trajectories[i] = [x_trajectories[i]; [real(u[n_states + n_excited + 1]) for u in sol_u]./k*1e3 ]
    y_trajectories[i] = [y_trajectories[i]; [real(u[n_states + n_excited + 2]) for u in sol_u]./k*1e3 ]
    z_trajectories[i] = [z_trajectories[i]; [real(u[n_states + n_excited + 3]) for u in sol_u]./k*1e3 ]
    
    times[i] = sol_t .* (1/Γ)
end

trajectories = [
    [[x[1],x[2],x[3]] for x ∈ zip(x_trajectories[i],y_trajectories[i],z_trajectories[i])] 
    for i ∈ 1:n_molecules
    ]
;
;
```

```julia
function trapped_molecules(x_trajectories, z_trajectories)
    survived = zeros(Bool, length(x_trajectories))
    for i ∈ eachindex(x_trajectories)
        survived[i] = (abs(x_trajectories[i][end]) <= 60e-3) && (abs(y_trajectories[i][end]) <= 60e-3) && (abs(z_trajectories[i][end]) <= 60e-3)
    end
    return survived
end
trapped = trapped_molecules(x_trajectories, z_trajectories)
;
```

```julia
lim = 0.2
plot()
for i ∈ (1:n_molecules)[trapped]
    plot!(times[i], x_trajectories[i], legend=nothing, ylim=(-lim, lim))
end
plot!(size=(1000,1000))
```
